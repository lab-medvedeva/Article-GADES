```python
import math

def read_memory_R(filename):
    df = pd.read_csv(filename, sep=' ')
    return df.iloc[0]['V1'] - df.iloc[0]['V2']


def read_memory_pythonic(filename):
    df = pd.read_csv(filename, sep=',')
    return np.abs(df.iloc[0]['result'] - df.iloc[0]['base'])
```

```python
import numpy as np
```

```python
results = []

for filename in glob.glob('../../Hobotnica-GPU/results/GeneratedDenseMemoryUsage/*_cells_*features/_*_*_memory.csv'):
    print(filename)
    group = pattern.match(filename)
    
    cells = int(group.group('cells'))
    features = int(group.group('features'))
    method = group.group('method')
    metric = group.group('metric')

    if method in ['CPU', 'factoextra', 'amap', 'GPU']:
        memory = read_memory_R(filename)
    else:
        memory = read_memory_pythonic(filename)
    
    # batch_size = int(group.group('batch_size'))
    #
#     print(cells, features, method, metric)
    df = pd.read_csv(filename)
    
    values = {
        'DataFile': filename,
        'metric': metric,
        'method': method,
        'wX': cells,
        'wY': features,
        'ID': f'{cells}_{features}',
        '|W|': cells * features,
        'logWX': np.log10(cells),
        'logWY': np.log10(features),
        'memory': memory,
        'logMemory': np.log10(memory),
        'Method': f'{metric}_{method}',
    }
        
    results.append(values)
```
